```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
my_img = cv2.imread('../Assets/penguen.jpg')
print(type(my_img))
```

```python
my_rgb_img = cv2.cvtColor(my_img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(my_rgb_img)
plt.axis('off')
```

```python
#flip fonksiyonu:grnty dikey,yatay veya her iki eksende evirmek iin kullanlr.
#say=0 ise grnt dikey eksende evrilir.
#say>0 ise grnt yatay eksende evrilir.
#kullanm : cv2.flip(grnt, say)
#tam say deeri = 0,1,-1 olabilir.

my_vertical_img = cv2.flip(my_rgb_img, 0) #dikey eksende evirme
my_horizontal_img = cv2.flip(my_rgb_img, 1) #yatay eksende evirme
my_both_img = cv2.flip(my_rgb_img, -1) #her iki eksende evirme
```

```python
fig = plt.figure(figsize=(10,7))
plt.subplot(2,2,1)
plt.imshow(my_rgb_img)
plt.title('Orijinal Grnt')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(my_vertical_img)
plt.title('Dikey evirme')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(my_horizontal_img)
plt.title('Yatay evirme')
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(my_both_img)
plt.title('Her ki Eksende evirme')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
#flip fonskiyonun ilevini manuel olarak tanmlamak:
my_flip_img = my_rgb_img[:,:,:] #satr,stun ve renk kanallarn al
my_flip_img = my_flip_img[:,::-1,:] #stunlar ters evir
plt.imshow(my_flip_img)
plt.axis('off')
```

```python

vertical = my_rgb_img[:,::-1,:]
horizanal = my_rgb_img[::-1,:,:]
both = my_rgb_img[::-1,::-1,:]

fig = plt.figure(figsize=(10,7))
plt.subplot(2,2,1)
plt.imshow(my_rgb_img)
plt.title('Orijinal Grnt')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(horizanal)
plt.title('Dikey evirme')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(vertical)
plt.title('Yatay evirme')
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(both)
plt.title('Her ki Eksende evirme')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
#kendi flip fonksiyonumuzu tanmlayalm:
def my_flip(image, direction):
    if direction == 0: #dikey eksende evirme
        image = image[:,::-1,:]
    elif direction == 1: #yatay eksende evirme
        image = image[::-1,:,:]
    elif direction == -1: #her iki eksende evirme
        image = image[::-1,::-1,:]
    else:
        raise ValueError("Ltfen doru deer giriniz: 0, 1 veya -1")
        
    return image

my_vertical_img2 = my_flip(my_rgb_img, 0) #dikey eksende evirme
my_horizontal_img2 = my_flip(my_rgb_img, 1) #yatay eksende
my_both_img2 = my_flip(my_rgb_img, -1) #her iki eksende evirme

fig = plt.figure(figsize=(10,7))
plt.subplot(2,2,1)
plt.imshow(my_rgb_img)
plt.title('Orijinal Grnt')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(my_vertical_img2)
plt.title('Yatay evirme')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(my_horizontal_img2)
plt.title('Dikey evirme')
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(my_both_img2)
plt.title('Her ki Eksende evirme')
plt.axis('off')
plt.tight_layout()
plt.show()
```
